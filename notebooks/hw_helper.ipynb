The Training pipeline loads the data from a single file. Modify the code to point the pipeline to a folder instead. The pipeline should load every file from the folder and combine them to create the dataset.

```python
# Split data/penguins.csv into data/data_folder/penguins1.csv and data/data_folder/penguins2.csv
import pandas as pd
import os
data = pd.read_csv('../data/penguins.csv')
data_folder = '../data/data_folder'
os.makedirs(data_folder, exist_ok=True)
data1 = data.sample(frac=0.5, random_state=42)
data2 = data.drop(data1.index)
data1.to_csv(os.path.join(data_folder, 'penguins1.csv'), index=False)
data2.to_csv(os.path.join(data_folder, 'penguins2.csv'), index=False)   
```
